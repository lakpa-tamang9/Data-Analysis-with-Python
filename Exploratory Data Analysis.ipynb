# Exploratory Data Analysis

Exploratory Data Analysis (EDA) is used by data analysts, and data scientists to analyse and examine data sets and summarise their key features. This is often done by employing data visualisation techniques. EDA helps to determine how data sources can be manipulated to provide the answers required, making it easier for data scientists to spot patterns, uncover anomalies, test hypotheses and confirm assumptions. The main purpose of EDA is to check the data before making hypotheses. EDA not only identifies obvious errors, but also allows us to better understand patterns in the data, detect outliers and unusual events, and find interesting relationships between variables.
In this tutorial, we use Python for EDA to identify missing values in a dataset, which is important to help us decide how to handle missing values for solving machine learning tasks. 

We can use EDA these to find shape of datasets, descriptive statistics of the dataset, correlation between the features in a dataset, acquiring more information about the data from each data frame, checking out the missing values in the data, encoding categorical features etc.

# Exampe 02: Now lets dive more into the EDA.

```python
import numpy as np
import os #Execute the functions provided by the operating system
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

path = os.path.join("housing-Copy1.csv") #can be downloaded from the link provided above.
housing = pd.read_csv(path)
housing.head()
```

```python
housing.info()
```

```python
housing.describe()
```

```python
#Now lets plot the data distribution of each features.
#For this we Create a histogram with 50 bars
#Histogram plot allows to discover and show the underlying frequency distribution of the data.
#This can be done to inspect the data to check its distribution, and finding the outliers.
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

In the next step we use scatter plotting to visualize the feature of the dataset. A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.1) # alpha : density measure
plt.show()
```

Scatter plotting can be done by color mapping the specific feature. For example, below we color map the population variable and show which location (latitude/longitude) consists of dense population. 

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.3,
    s=housing["population"]/100, label="population", figsize=(10,7),
    c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
    sharex=False) # s: size, c: color reference, cmap: mapping option
plt.legend()
plt.show()
```

Correlation: They can be used to investigate the dependence between multiple variables at the same time and to highlight the most correlated variables in a data table. In this visual, correlation coefficients are colored according to the value.
A correlation coefficient of 1 means that for every positive increase in one variable, there is a positive increase of a fixed proportion in the other. For example, shoe sizes go up in (almost) perfect correlation with foot length.

A correlation coefficient of -1 means that for every positive increase in one variable, there is a negative decrease of a fixed proportion in the other. For example, the amount of gas in a tank decreases in (almost) perfect correlation with speed.

Zero means that for every increase, there isn’t a positive or negative increase. The two just aren’t related.

#Pearson's correlation coefficient, 1: positive slope, -1: negative slope, 0: means that there is no data relationship, but that it is difficult to express linearly.

```python
corr_matrix = housing.corr()
print(corr_matrix["median_house_value"].sort_values(ascending=False)) # Sort in order of correlation coefficient
```

```python
print(corr_matrix["median_income"].sort_values(ascending=False)) 
```

Now lets create scatter matrix of all featurs of the data and see the realtionship between the variables. As we go from left to right in the scatter plot, the uphill pattern shows there is positive relation between two variables, and if opposite then there is negative relationship.

```python
#importing scatter_matric from pandas plotting library
from pandas.plotting import scatter_matrix
scatter_matrix(housing, figsize=(12, 8))
plt.show()
```

To plot the correlation map, we can use seaborn library from python. To install go to this link :https://seaborn.pydata.org/installing.html
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.       

```python
import seaborn as sns
plt.figure(figsize=(15,15))
sns.heatmap(corr_matrix, annot=True, fmt = '.2f', linewidths=.5, cmap='Blues')# annot: cell value display, fmt: number display format
```

The standard correlation with median_house_value was quite low, so it was not very helpful to view it as an independent variable.
However, from the above correlation plot we can see that there is a high correlation between these variables (total rooms, total bedrooms, population, and households), so now lets create more meaningful representations by using those variables.

```python
attributes = ["total_rooms", "households", "total_bedrooms", "population"] 
scatter_matrix(housing[attributes], figsize=(12, 8))
plt.show()
print(housing[attributes].corr())
```

Now lets create three new variables in our data and check their correlation. 

```python
# Create three new variables
housing["rooms_per_household"] = housing["total_rooms"]/housing["households"] #new variable 1
housing["bedrooms_per_room"] = housing["total_bedrooms"]/housing["total_rooms"] #new vairable 2
housing["population_per_household"]=housing["population"]/housing["households"] #new variable 3
corr_matrix = housing.corr()
print(corr_matrix["median_house_value"].sort_values(ascending=False)) #rooms_per_household , bedrooms_per_room has a greater correlation than the existing variable
```
